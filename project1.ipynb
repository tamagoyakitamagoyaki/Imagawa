```python
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas_datareader as pdr
import numpy as np

start_date = '1995-01-01'
end_date = '2022-01-01'

gdp_g = web.DataReader('CLVMNACSCAB1GQDE', 'fred', start_date, end_date)
log_gdp_g = np.log(gdp_g)

gdp_g_pct_change = gdp_g.pct_change(4)

cycle_g, trend_g = sm.tsa.filters.hpfilter(log_gdp_g, lamb=1600)

plt.plot(log_gdp_g, label="Original GDP (in log)")

plt.plot(trend_g, label="Trend")

plt.legend()
plt.show()
```

```python
plt.plot(cycle_g, label="Cyclical component")

plt.legend()
plt.show()

cycle_g_mean = cycle_g.mean()
cycle_g_std = cycle_g.std()

print("Cycle mean:", cycle_g_mean)
print("Cycle standard deviation:", cycle_g_std)
```
